# Inspect a next-scene prediction
Load the NPZ and JSON files emitted by `predict.py`.

```python
from pathlib import Path
import json
import numpy as np
import matplotlib.pyplot as plt

prefix = Path('../predictions/up_next')
summary = json.loads(prefix.with_suffix('.json').read_text())
with np.load(prefix.with_suffix('.npz')) as archive:
  probability_matrix = archive['probability_matrix']
summary['top_interactions'][:10]
```

```python
n = max(1, max([entry['character_i'] for entry in summary['top_interactions']] + [0]) + 1)
plt.figure(figsize=(7, 6))
plt.imshow(probability_matrix[:n, :n], vmin=0, vmax=1)
plt.colorbar(label='Predicted probability')
plt.xlabel('Character ID')
plt.ylabel('Character ID')
plt.title('Predicted next-scene interaction probabilities')
plt.show()
```
