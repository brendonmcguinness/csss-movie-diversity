# Inspect the processed movie dataset
This notebook summarizes scene density and character counts without loading all movies simultaneously.

```python
from pathlib import Path
import json
import numpy as np
import matplotlib.pyplot as plt

processed_dir = Path('../data/processed')
manifest = json.loads((processed_dir / 'manifest.json').read_text())
len(manifest['movies'])
```

```python
character_counts = [movie['n_characters'] for movie in manifest['movies']]
scene_counts = [movie['n_scenes'] for movie in manifest['movies']]
print('Character count range:', min(character_counts), max(character_counts))
print('Scene count range:', min(scene_counts), max(scene_counts))
```

```python
plt.figure(figsize=(7, 4))
plt.hist(character_counts, bins=20)
plt.xlabel('Speaking characters per movie')
plt.ylabel('Movies')
plt.title('Character-count distribution')
plt.show()
```

```python
record = manifest['movies'][0]
with np.load(processed_dir / record['data_path']) as archive:
  scenes = archive['scene_vectors']
density = scenes.mean(axis=1)
plt.figure(figsize=(9, 4))
plt.plot(density)
plt.xlabel('Scene index')
plt.ylabel('Packed adjacency density')
plt.title(record['title'])
plt.show()
```
